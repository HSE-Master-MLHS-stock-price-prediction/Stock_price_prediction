```python
import yfinance as yf
from matplotlib import pyplot as plt

from SingleTickerBacktesting import SimplePortfolioExperiment
```

```python
data = yf.Ticker('YNDX').history(period="max")
train_data = data[:1412]
experiment_data = data[1412:-250]
plt.figure(figsize = (18, 10))
plt.title('YNDX price dynamic from 01.01.2015', font = 'Times', fontsize = 16)
plt.grid()
plt.plot(data.index[:-250], list(train_data.Close) + [None]*len(experiment_data.Close))
plt.plot(data.index[:-250], [None]*len(train_data.Close) + list(experiment_data.Close))
plt.show()
```

```python
class NikiforovYNDXExperiment(SimplePortfolioExperiment):
    
    def make_action(self, ind):
        self.buy_stocks_per_money(self.free_money, ind, True)
```

```python
obj = NikiforovYNDXExperiment()
obj.start_experiment()
```

```python
plt.figure(figsize = (18, 10))
plt.grid()
plt.plot(obj.history_capital)
plt.show()
```

```python
plt.figure(figsize = (18, 10))
plt.grid()
plt.plot(obj.history_free_money)
plt.show()
```

```python
plt.figure(figsize = (18, 10))
plt.grid()
plt.plot(obj.histiry_stocks)
plt.show()
```
